```python
import subprocess

def getValue(s):
    return s.split()[-1]

def getLabel(s):
    return s.split()[0] + ' ' + s.split()[1]

for interval in range(0, 1000):
    second = interval * 0.47244
    stdoutdata = subprocess.getoutput("./sox target.wav _.wav trim " + str(second) + " 0.47244 stat")
    target_text = stdoutdata.splitlines()[8]
    target_value = getValue(target_text)
    target_label = getLabel(target_text)
    print(str(interval) + ',' + target_value)
```

```python
table_head = True
for interval in range(0, 1000):
    second = interval * 0.47244
    stdoutdata = subprocess.getoutput("./sox target.wav _.wav trim " + str(second) + " 0.47244 stat")
    target_text = stdoutdata.splitlines()[9]
    target_value = getValue(target_text)
    target_label = getLabel(target_text)
    if table_head:
        print('interval,'+'_'.join(target_label.split())[:-1])
        table_head = False
    print(str(interval) + ',' + target_value)
```
