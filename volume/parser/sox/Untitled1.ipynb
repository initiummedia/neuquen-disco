```python
import subprocess

def getValue(s):
    return s.split()[-1]

def getLabel(s):
    return s.split()[0] + ' ' + s.split()[1]

for second in range(0, 283+1):
    stdoutdata = subprocess.getoutput("./sox target.wav _.wav trim " + str(second) + " 1 stat")
    target_text = stdoutdata.splitlines()[8]
    target_value = getValue(target_text)
    target_label = getLabel(target_text)
    print(str(second) + ',' + target_value)
```
